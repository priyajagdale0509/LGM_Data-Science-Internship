Task 4 : Image to Pencil Sketch With Python


**import libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

 **Reading the image**

```python
img = cv2.imread("/content/Roses.jpg")
cv2_imshow(img)
```

**Converting image into GRAY SCALE IMAGE**




```python
#convert image from BGR to gray scale image 
gray_scale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale_image)
```

 **Inverting the gray scale image**

```python
#using bitwise_not functn to invert the image
inverted_img = cv2.bitwise_not(gray_scale_image)
cv2_imshow(inverted_img)
```

 **Image Smoothing**

```python
#Used gaussian blur technique with 21 x 21 pixel and 
#the default sigma values filter on the image to smmoothen our image
img_smoothing = cv2.GaussianBlur(inverted_img, (21, 21), sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

 **Final Pencil SKETCH**

```python
#Dividing the grayscale values of the image by the inverse of blurred image value which highlights the boldest edges of the image
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_scale_image, img_smoothing)
cv2_imshow(pencil_sketch)
```

 **Both image comparison**

```python
cv2_imshow(img)
cv2_imshow(pencil_sketch)
```
